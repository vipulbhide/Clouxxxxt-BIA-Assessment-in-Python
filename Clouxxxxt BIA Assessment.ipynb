```python
L606['Label_Container_to'] = L606['Column B Label Container To'].str[-5:]
```

```python
L606.head(10)
```

```python
del L606['Column B Label Container To']
```

```python
L606.columns = ['Desitnation_Zip', 'Label_Container_to']
```

```python
#Transposing all the zipcodes seperated by comman to assign the label container and store it in new dataframe

L606_1 = (L606.set_index(L606.columns.drop('Desitnation_Zip',601).tolist())
   .Desitnation_Zip.str.split(',', expand=True)
   .stack()
   .reset_index()
   .rename(columns={0:'Desitnation_Zip'})
   .loc[:, L606.columns]
   )
```

```python
L606_1.head(10)
```

```python
L606_1['Desitnation_Zip'] = L606_1['Desitnation_Zip'].str.replace(r'\t','')
```

```python
L606_1[['Source_Zip','Destination_Zip1']] = L606_1.Desitnation_Zip.str.split("-",expand=True)
```

```python
L606_1.head(10)
```

```python
L606_1.dtypes
```

```python
L606_1['Destination_Zip1'] = L606_1['Destination_Zip1'].fillna(0)


L606_1 = L606_1[['Source_Zip','Destination_Zip1','Label_Container_to']]

L606_1.head(10)
```

```python
L606_1['Source_Zip_int'] = L606_1['Source_Zip'].astype(str).astype(int)
L606_1['Destination_Zip1_int'] = L606_1['Destination_Zip1'].astype(str).astype(int)
L606_1.head(10)
```

```python
solution_df = {'Zip': [],
        'Label_Destination' : []
        }
final_df = pd.DataFrame(solution_df, columns = ['Zip', 'Label_Destination'])

for index,row in L606_1.iterrows():

   
    source_zip = row['Source_Zip_int']
    dest_zip = row['Destination_Zip1_int']
    key = row['Label_Container_to']
   

    if dest_zip - source_zip > 0 :
      final_dest_zip = dest_zip
    else:
      final_dest_zip = source_zip
    #print(final_dest_zip)
    
    while final_dest_zip >= source_zip:
          temp_df = pd.DataFrame([[source_zip,key]], columns=['Zip', 'Label_Destination'])
          final_df = final_df.append(temp_df)    
          source_zip += 1
```

```python
final_df.head(10)
```

```python
final_df['Zip'] = final_df['Zip'].apply(int).apply(str)
final_df['Zip_Updated'] = final_df['Zip'].apply(lambda x: x.zfill(5))
```

```python
final_df.head(10)
```

```python
final_df = final_df[['Zip_Updated','Label_Destination']]
final_df = final_df.rename(columns={'Label_Destination': 'Schemes'})
```

```python
final_df.head(10)
```

```python
Facility_Parcel_final_df = pd.merge(Facility_Parcel, final_df, left_on=['FACILITY_TRIM_ZIP'], right_on=['Zip_Updated'], how="left")
```

```python
Facility_Parcel_final_df['Schemes'] = Facility_Parcel_final_df['Schemes'].fillna('00000')
Facility_Parcel_final_df['Zip_Updated'] = Facility_Parcel_final_df['Zip_Updated'].fillna('00000')
Facility_Parcel_final_df.head(10)
```
